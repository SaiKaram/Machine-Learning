```python
#This is Logistic Regression Program for Multi Class Prediction

import sys
import copy
import numpy as np
from scipy.io import loadmat
from sklearn import metrics


#Hypothesis Function
def hypoF(theata,X):
    theataT = np.transpose(theata)
    xT = np.transpose(X)
    mul = np.matmul(theataT,xT)
    return 1 / (1 + np.exp(-mul))

#Gradient Descent Function
def gradDes(epsilon,alpha,theata,theataOld,X,y):
    hypoFunction = hypoF(theata,X)
    while(np.linalg.norm(theata - theataOld,ord=2) >= epsilon):
        theataOld = theata
        theata = theataOld - alpha * np.matmul(X.T,(np.array(hypoFunction)-y))
        hypoFunction = hypoF(theata,X)
    return theata


#Loading Data
dataTrain = loadmat('Train_Data.mat')
trainLabel = loadmat('Train_Label.mat')

dataTest =loadmat('Test_Data.mat')
testLabel =loadmat('Test_Label.mat')

X_train = dataTrain['trainData']
X_trainLabels = trainLabel['trainLabels']

X_test = dataTest['testData']
X_testLabels = testLabel['testLabels']

#Inserting the column of 1's
X_train = np.insert(X_train,0,1,axis=1)
X_test = np.insert(X_test,0,1,axis=1)

theata = np.ones(X_train[0].shape)
theataOld = theata * 999

X_trainLabels = [i[0] for i in X_trainLabels]
X_trainLabels = np.array(X_trainLabels)

#Dictionary stores the data sets for different labels
data = dict()

for j in range(10):
    labels = copy.deepcopy(X_trainLabels)
    
    for i in range(X_trainLabels.size):
        if(labels[i] == j):
            labels[i] = 1
            #print "labels = ",labels[i]
        else:
            labels[i] = 0
    
    index = str(j)
    data.update({index:labels})

#Error Tolerance
epsilon = 10
#Learning Rate
alpha = 0.02

#The list stores the Optimum Theata Values given by Gradient Descent Function
AllTheata = []

for i in range(10):
    index = str(i)
    AllTheata.append(gradDes(epsilon,alpha,theata,theataOld,X_train,data[index]))

# The list stores the probabilities obtained using the Optimum Theata Values
prediction = []




for i in range(len(AllTheata)):
    prediction.append(hypoF(AllTheata[i],X_test))
    
#The list stores the predicted labels    
maxPredictLabel = []


i = 0
for j in range(10000):
    maxi = max(prediction[i][j],prediction[i+1][j],prediction[i+2][j],prediction[i+3][j],prediction[i+4][j],prediction[i+5][j],prediction[i+6][j],prediction[i+7][j],prediction[i+8][j],prediction[i+9][j])
    if(maxi == prediction[i][j]):
        maxPredictLabel.append(i)
    elif(maxi == prediction[i+1][j]):
        maxPredictLabel.append(i+1)
    elif(maxi == prediction[i+2][j]):
        maxPredictLabel.append(i+2)
    elif(maxi == prediction[i+3][j]):
        maxPredictLabel.append(i+3)
    elif(maxi == prediction[i+4][j]):
        maxPredictLabel.append(i+4)
    elif(maxi == prediction[i+5][j]):
        maxPredictLabel.append(i+5)
    elif(maxi == prediction[i+6][j]):
        maxPredictLabel.append(i+6)
    elif(maxi == prediction[i+7][j]):
        maxPredictLabel.append(i+7)
    elif(maxi == prediction[i+8][j]):
        maxPredictLabel.append(i+8)
    elif(maxi == prediction[i+9][j]):
        maxPredictLabel.append(i+9)



count = 0
for i in range(len(maxPredictLabel)):
    if(maxPredictLabel[i] == X_testLabels[i]):
        count += 1
        
 

accuracy = count / float(X_testLabels.size)

print "The model gives the accuracy :",accuracy
```
